```python
#===========================================================================
# Name      : AS4_wildlife_start.py
# Purpose   : Display lists of wildlife observations and allow for entry
#             of new observations
# Written By: J.Verran Nov. 2021
#
# Wildlife.txt processing   ( wildlife.txt )
#===========================================================================


#===========================================================================
# Function: menu()
# Purpose : Display the menu of options and request a choice from the user.
#===========================================================================
def menu():

    # Present the menu until a valid choice is entered.
    choice = 0
    while choice < 1 or choice > 7:
        print("\n\n\t Wildlife information Management")
        print("\t======== Selection Menu =========")
        print("\t=================================")
        print("\t1. Show all Observations")
        print("\t2. Show Observations by a Person")
        print("\t3. Show Observations of a Species")
        print("\t4. Add an Observation")
        print("\t5. Add an Observer")
        print("\t6. Add a Species")
        print("\t7. Exit")

        # A non-number will be set to zero.  (Causing the menu to show again)
        try:
            choice = input("\n\tEnter your selection: ")
            if choice == "":
                choice = 0
            else:
                choice = int(choice)
        except:
            print("\n\t%s is not valid\n" % choice)
            choice = 0

    # Return the number entered.
    return choice


    #===========================================================================
    # Function: display()
    #===========================================================================
    ############################################################################
    ## Write three separate functions:
    ##      1. to display all entries
    ##      2. to display entries for a selected observer
    ##      3. to display entries for a selected species
    ##
    ## Or write one function that works for all three options.
    ##   - Hint: For this option, use function arguments to indicate which
    ##           option was selected, and the values to list.
    ############################################################################
    def display(command,choice):

        observations = []  
        infile = open("wildlife.txt", "r")
        line = infile.readline()

        while line != "":
            line = line.rstrip('\n')
            observations.append(line.split(","))
            line = infile.readline()
        infile.close()

        observations[5][2] = "Dec. 19"
        observations[5].insert(3," "+str(2021))

        for i in range(len(observations)):
            #print(observations[i])
            observations[i][2] = (observations[i][2].strip('"').capitalize()+","+observations[i][3])
            observations[i].remove(observations[i][3])


        def print_header():
            print("%-10s \t %-10s \t %-10s \t %-10s \t %-10s \t %-10s \t %-10s"
                      %("Observer","Time","Date","Town","Lat","Long","Species"))
            print("   "+"="*110)



        def print_all_entries(observations=observations):
            for observation in observations:
                print("%-10s \t %-10s \t %-10s \t %-10s \t %-10s \t %-10s \t %-10s" %
                                                  (observation[0].strip('"'),
                                                   observation[1].strip('"'),
                                                   observation[2].strip('"'),
                                                   observation[3].strip('"'),
                                                   observation[4].strip('"'),
                                                   observation[5].strip('"'),
                                                   observation[6].strip('"')))


        def print_by_observer(person_index = choice, observations = observations):
            for observation in observations:
                if observation[0].strip('"').upper() == names[choice]:
                    print("%-10s \t %-10s \t %-10s \t %-10s \t %-10s \t %-10s \t %-10s" %
                                                      (observation[0].strip('"'),
                                                       observation[1].strip('"'),
                                                       observation[2].strip('"'),
                                                       observation[3].strip('"'),
                                                       observation[4].strip('"'),
                                                       observation[5].strip('"'),
                                                       observation[6].strip('"')))

        def print_by_species(species_index = choice, observations = observations):
            for observation in observations:
                if observation[6].strip('"').upper() == species[choice]:
                    print("%-10s \t %-10s \t %-10s \t %-10s \t %-10s \t %-10s \t %-10s" %
                                                      (observation[0].strip('"'),
                                                       observation[1].strip('"'),
                                                       observation[2].strip('"'),
                                                       observation[3].strip('"'),
                                                       observation[4].strip('"'),
                                                       observation[5].strip('"'),
                                                       observation[6].strip('"')))

        if command == 1:
            print_header()
            print_all_entries()

        elif command ==2:
            print_header()
            print_by_observer()

        elif command ==3:
            print_header()
            print_by_species()
        



    #===========================================================================
    # Function: load_names()
    # Purpose : Populate the names list with all the people in the file
    #===========================================================================
    def load_names():
        global names

        # If the list of names is already populated, the try: action will succeed
        # and the names will not be read again.
        # If the names haven't been loaded yet, the exception will be run to load
        # the names from the file.
        try:
            x = names[0]
        except:
            names = []

            infile = open("wildlife.txt","r")
            for data in infile:
                f = data.split(",")                # f is all the fields
                name = f[0].strip('"')             # f[0] is the name
                if name.upper() not in names:      # if not in the list append it
                    names.append(name.upper())

            infile.close()
            
    

    #===========================================================================
    # Function: get_person()
    # Purpose : Show the names and get a selection from the list.
    #===========================================================================
    def get_person():

        load_names()                               # Build the list of names

        print("\n\tThese are the current names")
        print("\t===========================")

        # Show the names each with a number (1 greater than the index)
        for name in names:
            print("\t\t%d. %s" % (names.index(name)+1,name.capitalize()))

        # Get a number within the valid range
        n = 0
        while n < 1 or n > len(names):
            try:
                n = input("\n\tChoose a number : ")    
                n = int(n)

                # Force the exception if the number is not out of range.

                if n < 0 or n > len(names):
                    x = int("A")
            except:
                print("Choose from 1 to %d" % (len(names)))

        # Return the index number for the name selected
        # - (1 less than the number entered.)

        return n - 1
    
    

    #===========================================================================
    # Function: load_species()
    # Purpose : Populate the list of species
    #===========================================================================
    def load_species():
        global species

        # Similar to load_names(), the exception will populate the list of species
        try:
            x = species[0]
        except:
            species = []

            infile = open("wildlife.txt","r")

            for data in infile:
                data = data.replace('"','')
                data = data.strip('\n')
                f = data.split(",")                # f is all the fields       
                sp = f[len(f)-1]                   # The last field is the species
                if sp.upper() not in species:      # if not in the list append it
                    species.append(sp.upper())

            infile.close()
    
    #===========================================================================
    # Function: get_species()
    # Purpose : Display the species and get a selection.
    #===========================================================================
    def get_species():

        load_species()                             # Build the list of species

        print("\n\tThese are the current Species")
        print("\t=============================")

        # Show the species

        for sp in species:
            print("\t\t%d. %s" % (species.index(sp)+1,sp.capitalize()))

        # Repeat until a valid number is entered.

        n = 0
        while n < 1 or n > len(species):
            try:
                n = input("\n\tChoose a number : ")      
                n = int(n)

                # If out of range, force an error to show the exception.

                if n < 0 or n > len(species):
                    x = int("A")
            except:
                print("Choose from 1 to %d" % (len(species)))

        # Return the index for the selected species.

        return n - 1
    
    
    ############################################################################
    ## Complete the function below to add a new observation
    ############################################################################
    #===========================================================================
    # Function: add_observation()
    # Purpose : Add an observation entry to the file.
    #===========================================================================
    def add_observation():

        def persons_name():
            name_in = ""
            while name_in == "":
                name_in = input("Please enter a name: ")
                name_in = name_in.lstrip(" ")
                if (name_in.isnumeric() == True) or (name_in == "") or (name_in.isspace == True) :
                    print("Please enter a valid name with only characters.")
                    name_in = ""
                elif name_in.isnumeric() == False:
                    print("The name you entered is: %5s" %(name_in))
                    return name_in.capitalize()

        def time_input():

            hour_in = ""
            hours = [str(hrs) for hrs in range(1,13)]

            while hour_in == "":
                hour_in = input("Please enter the hour of observation from 1-12: ")
                hour_in = hour_in.strip(" ")

                if (hour_in.isnumeric() == False) or (hour_in not in hours):
                    print("Please enter an acceptable value in the range of 1-12.")
                    hour_in = ""
                elif hour_in in hours:
                    print("You selected hour value: %s" %(hour_in))


            minute_in = ""
            minutes = [str(mins) for mins in range(0,60)]

            while minute_in == "":
                minute_in = input("Please enter the minute of observation from 0-59: ")
                minute_in.strip(" ")

                if (minute_in.isnumeric() == False) or (minute_in not in minutes):
                    print("Please enter an acceptable value in the range of 0-59.")
                    minute_in = ""

                elif minute_in in minutes:
                    print("You selected minute value: %s" %(minute_in))

            the_time = hour_in + ":" + minute_in                

            return the_time


        def date_input():


            def select_month():
                month_selected = ""
                months = ["Jan.", 
                      "Feb.",
                      "Mar.",
                      "Apr.",
                      "May.",
                      "Jun.",
                      "Jul.",
                      "Aug.",
                      "Sep.",
                      "Oct.",
                      "Nov.",
                      "Dec."]

                # Present the menu until a valid choice is entered.
                choice = 0
                while choice < 1 or choice > 12:
                    print("\n\n\t Select Month of Observation")
                    print("\t======== Selection Menu =========")
                    print("\t=================================")
                    print("\t1. January")
                    print("\t2. February")
                    print("\t3. March")
                    print("\t4. April")
                    print("\t5. May")
                    print("\t6. June")
                    print("\t7. July")
                    print("\t8. August")
                    print("\t9. September")
                    print("\t10. October")
                    print("\t11. November")
                    print("\t12. December")

                    # A non-number will be set to zero.  (Causing the menu to show again)
                    try:
                        choice = input("\n\tEnter your selection: ")
                        if choice == "":
                            choice = 0
                        else:
                            choice = int(choice)
                    except:
                        print("\n\t%s is not valid\n" % choice)
                        choice = 0

                    # Return the number entered.
                month_selected = months[choice - 1]

                return month_selected

            def select_day(month):

                jan_days = [str(day) for day in range(1,32)]
                feb_days = [str(day) for day in range(1,30)]
                mar_days = [str(day) for day in range(1,32)]
                apr_days = [str(day) for day in range(1,31)]
                may_days = [str(day) for day in range(1,32)]
                jun_days = [str(day) for day in range(1,31)]
                jul_days = [str(day) for day in range(1,32)]
                aug_days = [str(day) for day in range(1,32)]
                sep_days = [str(day) for day in range(1,31)]
                oct_days = [str(day) for day in range(1,32)]
                nov_days = [str(day) for day in range(1,31)]
                dec_days = [str(day) for day in range(1,32)]

                if month == "Jan.":
                    month_selected = jan_days
                elif month == "Feb.":
                    month_selected = feb_days
                elif month == "Mar.":
                    month_selected = mar_days
                elif month == "Apr.":
                    month_selected = apr_days
                elif month == "May.":
                    month_selected = may_days
                elif month == "Jun.":
                    month_selected = jun_days
                elif month == "Jul.":
                    month_selected = jul_days
                elif month == "Aug.":
                    month_selected = aug_days
                elif month == "Sep.":
                    month_selected = sep_days
                elif month == "Oct.":
                    month_selected = oct_days
                elif month == "Nov.":
                    month_selected = nov_days
                elif month == "Dec.":
                    month_selected = dec_days
4
                day_selected = ""
                while day_selected == "":
                    day_selected = input("For the month of %s please enter the day of observation from %s - %s: " %(month,month_selected[0],month_selected[-1]))
                    day_selected = day_selected.strip(" ")

                    if (day_selected.isnumeric() == False) or (day_selected not in month_selected):
                        print("Please enter an appropriate value from %s - %s for the month of %s." %(month_selected[0],month_selected[-1],month))
                        day_selected = ""
                    elif day_selected in month_selected:
                        print("You selected the day %s in the month of %s" %(day_selected,month))
                return day_selected

            def select_year():
                years = [str(year) for year in range(1900,2022)]

                year_selected = ""
                while year_selected == "":
                    year_selected = input("Please enter a year for the observation: ")
                    year_selected = year_selected.strip(" ")

                    if (year_selected.isnumeric() == False) or (year_selected not in years):
                        print("Please enter an appropriate year of observation in the range %s - %s" %(years[0],years[-1]))
                        year_selected = ""
                    elif year_selected in years:
                        print("You selected the year of observation to be: %s" %(year_selected))

                return year_selected



            ####Return Date String by Calling Functions!#####

            month_selected = select_month()
            day_selected = select_day(month_selected)
            year_selected = select_year()

            date_string = month_selected+" " + day_selected+", "+year_selected


            return date_string


            def select_location():
                location_in = ""
                while location_in == "":
                    location_in = input("Please enter a location for the observation: ")
                    location_in = location_in.strip(" ")
                    if (location_in.isnumeric() == True) or (location_in == "") or (location_in.isspace == True) :
                        print("Please enter a valid locatioon name with only characters.")
                        location_in = ""
                    elif location_in.isnumeric() == False:

                        print("The location you entered is: %5s" %(location_in))
                        return location_in.capitalize()



            def get_lat_long():
                lat_selection=-1000
                while lat_selection==-1000:
                    try:
                        lat_selection = input("Please enter a latitude for the observation between [-90,+90]: ")
                        #First check to see if the answer is numerical
                        if type(float(lat_selection)) != float:
                            print("Please enter a latitude for the observation between [-90,+90]: ")    
                            raise ValueError()


                        lat_selection = float(lat_selection)    
                        if (lat_selection <= 90) and (lat_selection >= -90):
                            print(f"You have selected latitude: {lat_selection}")    
                            break
                        else:

                            raise ValueError()

                    except ValueError:      
                        print("Please enter a latitude for the observation between [-90,+90]")
                        lat_selection=-1000

                long_selection=-1000
                while long_selection==-1000:
                    try:
                        long_selection = input("Please enter a longitude for the observation between [-180,+180]: ")
                        #First check to see if the answer is numerical
                        if type(float(long_selection)) != float:
                            print("Please enter a longitude for the observation between [-180,+180]: ")    
                            raise ValueError()


                        long_selection = float(long_selection)    
                        if (long_selection <= 180) and (long_selection >= -180):
                            print(f"You have selected longitude: {long_selection}")    
                            break
                        else:

                            raise ValueError()

                    except ValueError:      
                        print("Please enter a longitude for the observation between [-180,+180]")
                        long_selection=-1000

                return lat_selection, long_selection

            def insert_species():
                species_in = ""
                while species_in == "":
                    species_in = input("Please enter the observed species: ")
                    species_in = species_in.strip(" ")
                    if (species_in.isnumeric() == True) or (species_in == "") or (species_in.isspace == True) :
                        print("Please enter a valid locatioon name with only characters.")
                        species_in = ""
                    elif species_in.isnumeric() == False:

                        print("The species you entered is: %5s" %(species_in))
                        return species_in






        name_in = persons_name()
        time_in = time_input()
        date_in = date_input()
        location_in = select_location()
        lat_in ,long_in = get_lat_long()
        species_in = insert_species()

        file_object = open('wildlife.txt','a')

        file_object.write(name_in+","+time_in+","+date_in+","+location_in+","+str(lat_in)+","+str(long_in)+","+species_in+"\n")

        file_object.close()
        
        
    
    
    #===========================================================================
    # Function: add_observer()
    # Purpose : Add a name to the observer list.
    #===========================================================================
    def add_observer():
        load_names()                               # Build the list of names

        print("\n   Current Observer Names")
        print("   "+"="*22)


        for name in names:
            print("\t%s" % name)
        print("")

        # Prevent entry of an existing name or an empty string

        while name.upper() in names or name == "":
            name = input("Enter the new name: ")     
            if name.upper() in names:
                print(f"\n\t{name.capitalize()} is already in the Observer list\n")

        # Add the name to the list of observers.

        names.append(name.upper())
        

    #===========================================================================
    # Function: add_species()
    # Purpose : Add a species to the species list
    #===========================================================================
    def add_species():
        load_species()                               # Build the list of species

        print("\n   Current Species Observed")
        print("   "+"="*24)


        for sp in species:
            print("\t%s" % sp)
        print("")

        # Get the species and append to the list
        # Prevent entry of an existing species or an empty string

        while sp.upper() in species or sp == "":
            sp = input("Enter the new species: ")      
            if sp.upper() in species:
                print(f"\n\t{sp.capitalize()} is already in the Species list\n")

        # Add the species to the list of animal types.

        species.append(sp.upper())
        
        
       
    #===========================================================================
    # Function: main()
    # Purpose : Get the selection from the menu and call the function associated
    #           with the choice
    #===========================================================================
    def main():

        # Call the menu function to show the options
        command = 1
        while command > 0 and command < 7:

            # Get the selection from the menu and call the corresponding function
            command = menu()


            ###################################################################
            # Assignment:
            #    - Complete the actions for selections 1, 2, 3, and 4
            ###################################################################
            if command == 1:
                # Call function to display all observations
                print("\nCall option 1 function: Display all observations")
                display(command, choice=0)

            elif command == 2:
                # Call function to: Select a person --> get_person()
                choice = get_person()
                # Call function to: Display observations for the person
                print(f"\nCall option 2 function: Display for a person: {names[choice]}")
                display(command,choice)

            elif command == 3:
                # Call function to: Select a species
                choice = get_species()
                # Call function to: Display observations of a species
                print(f"\nCall option 3 function: Display for a species: {species[choice]}")
                display(command,choice)

            elif command == 4:
                # Call function to: Add a new observation entry
                print("\nCall option 4 function: Add and observation")
                add_observation()

            elif command == 5:
                add_observer()             # Add a person to the list

            elif command == 6:
                add_species()              # Add a species to the list

#===========================================================================
# Mainline:
# Purpose : Start the process by callin the main function
#===========================================================================
main()


        
    
```
