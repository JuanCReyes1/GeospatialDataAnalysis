```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import pandas as pd

from sklearn.cluster import DBSCAN
```

```python
def show_clusters(X,cluster):
    df = pd.DataFrame(dict(x=X[:,0],y=X[:,1],label=cluster))
    colors = {-1:"red",0:"blue",1:"orange",2:"green",3:"skyblue"}
    fig, ax = plt.subplots(figsize=(8,8))
    grouped = df.groupby("label")

    for key, group in grouped:
        group.plot(ax=ax,kind='scatter',x='x',y='y',label=key,color=colors[key])
    plt.xlabel("X_1")
    plt.ylabel("X_2")
    plt.show()
```

# Create dataset

```python
X,y = datasets.make_blobs(n_samples = 20000, centers=4, n_features=2, random_state=20,cluster_std=0.5)

df = pd.DataFrame(dict(x=X[:,0],y=X[:,1]))
```

```python
fig, ax = plt.subplots(figsize=(8,8))
df.plot(ax=ax, kind = 'scatter', x='x',y='y')

plt.xlabel('X_1')
plt.xlabel('X_2')
plt.show()
```

```python
varied_X,varied_y = datasets.make_blobs(n_samples=2000, centers=3, cluster_std=[1.0, 2.5, 0.5], random_state=10)

transformation = [[0.6, -0.6], [-0.4, 0.8]]
X_aniso = np.dot(varied_X, transformation)
aniso = (X_aniso, y)

df_varied = pd.DataFrame(dict(x=X_aniso[:,0],y=X_aniso[:,1]))
```

```python
fig, ax = plt.subplots(figsize=(8,8))
df_varied.plot(ax=ax, kind = 'scatter', x='x',y='y')

plt.xlabel('X_1')
plt.xlabel('X_2')
plt.show()
```

The two most important parameters for DBSCAN are the epsilon and min_samples parameters which will parametrize the final amount of clusters in our resultant set. 

```python
eps = 1
samples = 2

clustering = DBSCAN(eps=eps,min_samples=samples).fit(varied_X)

cluster = clustering.labels_
#len(set(cluster))
show_clusters(varied_X,cluster)
```

```python
n_samples = 500
noisy_circles_X,noisy_circles_y = datasets.make_circles(n_samples=n_samples, factor=0.4, noise=0.05)

noisy_circles_df = pd.DataFrame(dict(x=noisy_circles_X[:,0],y=noisy_circles_X[:,1]))
```

```python
fig, ax = plt.subplots(figsize=(8,8))
noisy_circles_df.plot(ax=ax, kind = 'scatter', x='x',y='y')

plt.xlabel('X_1')
plt.xlabel('X_2')
plt.show()
```

```python
eps = 0.11
samples = 2

clustering = DBSCAN(eps=eps,min_samples=samples).fit(noisy_circles_X)

cluster = clustering.labels_
#len(set(cluster))
show_clusters(noisy_circles_X,cluster)
```

```python
noisy_moons_X,noisy_moons_y = datasets.make_moons(n_samples=n_samples, noise=0.05)

noisy_moons_df = pd.DataFrame(dict(x=noisy_moons_X[:,0],y=noisy_moons_X[:,1]))
```

```python
fig, ax = plt.subplots(figsize=(8,8))
noisy_moons_df.plot(ax=ax, kind = 'scatter', x='x',y='y')

plt.xlabel('X_1')
plt.xlabel('X_2')
plt.show()
```

```python
eps = 0.11
samples = 2

clustering = DBSCAN(eps=eps,min_samples=samples).fit(noisy_moons_X)

cluster = clustering.labels_
#len(set(cluster))
show_clusters(noisy_moons_X,cluster)
```
